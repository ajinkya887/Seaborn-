```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

#linePlot

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,5,9,2,1,3]
```

```python
plt.plot(var, var_1)
plt.show()
```

```python
#using seaborn
#in seaborn we have to give dataframe

x_1 = pd.DataFrame({"var":var, "var_1":var_1})

sns.lineplot(x = "var", y = "var_1", data = x_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x ="bill_length_mm", y = "flipper_length_mm", data = y_1)
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm", y = "flipper_length_mm", data = y_1, hue = "sex", palette = 'Accent')
plt.show()
#'hue' is used to visualize the data of different categories in one plot
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=y_1, hue="sex", palette='Accent', style="sex", markers=["o", ">"])
plt.show()
```

#Bar Plot

```python
sns.barplot(x=y_1.island, y = y_1.bill_length_mm) 
plt.show()
```

```python
sns.barplot(x="island", y = "bill_length_mm", data = y_1) 
plt.show()
```

```python
sns.barplot(x="island", y = "bill_length_mm", data = y_1, hue = "sex") 
plt.show()
```

```python
#Set order of parameters
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y = "bill_length_mm", data = y_1, hue = "sex", order = order_1) 
plt.show()
```

```python
#hue Order
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y = "bill_length_mm", data = y_1, hue = "sex", order = order_1, hue_order = ["Female", "Male"]) 
plt.show()
```

```python
#horizontal graph
#to make horizontal graph both the parameters x and y should be in numeric form
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y = "bill_length_mm", data = y_1, orient="h") 
plt.show()
```

```python
sns.barplot(x="island", y = "bill_length_mm", data = y_1, color = "y") 
plt.show()
```

#Histogram plot

```python
sns.displot(y_1["flipper_length_mm"])
plt.show()
```

#Count Plot

```python
#what is difference between countplot and barplot?
#countplot plots the count of the number of records by category.
# Barplot plots a value or metric for each category (by default, baplot plots the mean of a variable, by category)
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x = "sex", data = var)
plt.show()
```

```python
sns.countplot(x = "sex", data = var, hue = "smoker")
plt.show()
```
